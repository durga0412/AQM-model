```python
import numpy as np
import pandas as pd
import time
from math import sqrt
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.activations import *
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
#from keras import *
import warnings
warnings.filterwarnings('ignore')


from firebase import firebase
import time
import pandas as pd
```

```python
firebase = firebase.FirebaseApplication("https://aqmnode-d979d.firebaseio.com/", None)
while True:
    AQI = firebase.get('/AQI', None)
    print("AQI:", AQI)

    Humidity = firebase.get('/Humidity', None)
    print("Humidity",Humidity)
    TEMP = firebase.get('/Temp', None)
    print("Temp",TEMP)
    WD = 0
    WS = 0
    BP = 0
    SR = 0
    print("WD", WD)
    print("WS", WS)
    print("BP", BP)
    print("SR", SR)


    data = [{'AQI': AQI, 'AT': TEMP, 'BP': BP, 'RH': Humidity,'SR': SR, 'WD': WD, "WS": WS}]
    print(data)
    df = pd.DataFrame(data)
    df.to_csv("firebase_base.csv")




    ####change the path here
    model = load_model("LSTM_one_value.h5py")
    #model.summary()


    ###Change the path here
    df = pd.read_csv("firebase_base.csv")
    df1 = df.drop(columns=['Unnamed: 0'])
    df1.head()



    df1.to_csv("LSTM_Data.csv")

    values = df1.values
    #print(df1.head())
    #print(values)

    values = values.astype('float32')
    # normalize features
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(values)



    reframed = pd.DataFrame(scaled)

    #print(reframed.head())

    values = reframed.values
    test = values
    #print(test)
    test_X = test
    # reshape input to be 3D [samples, timesteps, features]
    test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
    #print(test_X.shape)


    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))

    # invert scaling for forecast
    inv_yhat = np.concatenate((yhat, test_X[:, 1:]), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat)
    inv_yhat = inv_yhat[:,0]
    print("Predicted_AQI:",inv_yhat)

    firebase.post('predAQI_log', str(inv_yhat))
    firebase.put('/','PredAQI', str(inv_yhat))

    time.sleep(100)



```
